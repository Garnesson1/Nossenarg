```python
data = pd.read_excel('Data Set for Johan.xlsx', sheet_name = 'GDP YoY% (raw)')
urb = pd.read_excel('Data Set for Johan.xlsx', sheet_name = 'Urbanisation (raw)')
cap =  pd.read_excel('Data Set for Johan.xlsx', sheet_name = 'GDP per capita PPP USD (raw)')
pop =  pd.read_excel('Data Set for Johan.xlsx', sheet_name = 'Population (normalised)')
```

```python
data_ind = data.loc[0:6]
data_raw = data.loc[7:]

urb_ind = urb.loc[0:6]
urb_raw = urb.loc[7:]

cap_ind = cap.loc[0:6]
cap_raw = cap.loc[7:]
```

```python
dat1 = pd.DataFrame()
dat2 = pd.DataFrame()
def melter(data1, data2): 
    global Final
    dat1 = pd.melt(frame = data1, id_vars = ['YEAR'])
    dat2 = pd.melt(frame = data2, id_vars = ['YEAR'])
    datfin = dat2.pivot(index = 'variable', columns = 'YEAR', values = 'value')
    Final = datfin.join(dat1.set_index('variable'), on = 'variable')
    Final['Year'] = Final.index
    return Final
    
```

```python
melter(data_raw,data_ind)
Final_GDP = Final
```

```python
melter(urb_raw, urb_ind)
Final_URB = Final
```

```python
melter(cap_raw, cap_ind)
Final_CAP = Final
```

```python
Final_GDP.columns = [['DataFamily','DataUnit','Indicator','MetricNature','Period','Product','Source','Country','GDP', 'Year' ]]
Final_URB.columns = [['DataFamily','DataUnit','Indicator','MetricNature','Period','Product','Source','Country','Urbanisation', 'Year' ]]
Final_CAP.columns = [['DataFamily','DataUnit','Indicator','MetricNature','Period','Product','Source','Country','GDP/Capita', 'Year' ]]
```
